```python
# Create new feature FamilySize to count total Parch and SibSp + 1
# create new feature IsAlone with binary value to know the person alone or not
# create new feature Title from split in column Name to get title
# create new feature FareBin with cut fare by quartile
# create new feature AgeBin with cut Age by sorting data dan divide to 5 group
```

```python
for dataset in data_clean:
    dataset['FamilySize'] = dataset['Parch'] + dataset['SibSp'] + 1
    dataset['IsAlone'] = 1 
    dataset['IsAlone'].loc[dataset['FamilySize']>1] = 0
    dataset['Title'] = dataset['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4)
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)
```
